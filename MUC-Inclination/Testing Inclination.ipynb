```python
# this function creates all in-line plots. DO NOT DELETE.
import matplotlib
matplotlib.use('TkAgg')
%matplotlib inline
import matplotlib.pyplot as plt


def plot_data(data):
    # adjust rcParams as needed to make your graph readable
    plt.rcParams['figure.figsize'] = [10, 5]
    plt.title('X,Y,Z Plot')
    plt.plot([row[0] for row in data],[row[1] for row in data], label="x")
    plt.plot([row[0] for row in data],[row[2] for row in data], label="y")
    plt.plot([row[0] for row in data],[row[3] for row in data], label="z")
    plt.legend(loc='upper right')
    plt.show()
```

```python
# acceleration-withg-backfront

# this file is test.py
import parser_data
# numpy imported to check if the student has numpy
import numpy


data_with_gravity_backfront = parser_data.get_data("acceleration-withg-backfront.csv")
plot_data(data_with_gravity_backfront)
```

```python
# θ = arctan(Az / sqrt(Ax^2 + Ay^2))

import math
import matplotlib.pyplot as plt
import numpy as np

list_of_angles = []
list_of_time = []
time = 0
for list_accelerometer in data_with_gravity_backfront:
    time = list_accelerometer[0]
    x = list_accelerometer[1]
    y = list_accelerometer[2]
    z = list_accelerometer[3]
#     angle_radians = math.atan(z / (math.sqrt((math.pow(x, 2)) + (math.pow(y, 2)))))
    angle_radians = math.atan(math.sqrt(math.pow(x, 2) + math.pow(y, 2)) / z)
    angle_degrees = angle_radians * (180/math.pi)
    list_of_angles.append(angle_degrees)
    list_of_time.append(time)
    
    
# print(list_of_angles)
```

```python
plt.plot(list_of_time, list_of_angles)  
plt.xlabel('time') 
plt.ylabel('angle') 
  
# displaying the title
plt.title("time & angle of inclination WITH GRAVITY")
plt.show()  
```
