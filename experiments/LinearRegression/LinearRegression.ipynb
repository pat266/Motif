```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x_data = [0.0005369690916273328, 0.000390707736545139, 0.0004271815617879224, 0.000441518274943034, 0.0003724842495388455]
x_train = np.array(x_data).reshape((-1, 1))
y_train = [0, 271, 154, 38, 382]
```

```python
lr = LinearRegression().fit(x_train, y_train)
```

```python
r_sq = lr.score(x_train, y_train)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {lr.intercept_}")
print(f"coefficients: {lr.coef_}")
```

```python
# predict the weight
lr.predict(np.array([0.00042342232132]).reshape(-1, 1))
```

```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, lr.predict(x_train), color = "green")
plt.title("Weight vs Intesity (Training set)")
plt.xlabel("Relative Intensity")
plt.ylabel("Weight")
plt.show()
```
